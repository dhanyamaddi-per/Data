```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Earth=pd.read_pickle("cleaned_earthquakes.pkl")
```

```python
plt.figure(figsize=(12, 6))
top_places = Earth['place'].value_counts().nlargest(10)
sns.barplot(x=top_places.values, y=top_places.index, palette="viridis")
plt.title("Top 10 Locations by Earthquake Count")
plt.xlabel("Number of Earthquakes")
plt.ylabel("Location")
plt.show()
```

```python
Earth['time'] = pd.to_datetime(Earth['time'])
Earth['month'] = Earth['time'].dt.to_period('M')

monthly_counts = Earth['month'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
monthly_counts.plot(kind='line', marker='o')
plt.title("Monthly Earthquake Counts (2023)")
plt.xlabel("Month")
plt.ylabel("Number of Earthquakes")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(Earth['mag'], Earth['depth'], alpha=0.4, s=8)
plt.gca().invert_yaxis()  # depth increases downward
plt.xlabel('Magnitude')
plt.ylabel('Depth (km)')
plt.title('Magnitude vs. Depth')
plt.tight_layout()
plt.show()
```
